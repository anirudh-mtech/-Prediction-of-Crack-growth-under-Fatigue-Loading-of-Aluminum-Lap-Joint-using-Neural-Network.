Creating the neural model using Sequential API

```python
from keras import layers
from keras import Sequential
```

```python
model=Sequential([
    layers.Dense(40,activation="linear",input_shape=(3,),dtype=np.float64,name="layer1"),
    layers.Dense(80,activation="relu",dtype=np.float64,name="layer2"),
    layers.Dense(1,name="output",dtype=np.float64)
])
model.build()
```

```python
model.summary()
```

```python
model.compile(optimizer="Adam",loss=[keras.losses.MeanSquaredError()],metrics=[keras.metrics.MeanSquaredError()])
```
